```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
sns.set(style="whitegrid")
```

```python
# Bar chart:
plt.figure(figsize=(6, 4))
df['Pclass'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Histogram:
plt.figure(figsize=(6, 4))
df['Age'].dropna().plot(kind='hist', bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot:
plt.figure(figsize=(6, 4))
plt.scatter(df['Age'], df['Fare'], alpha=0.5, color='green')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.tight_layout()
plt.show()
```

```python
# Box plot:
plt.figure(figsize=(6, 4))
sns.boxplot(x='Pclass', y='Age', data=df, palette='Set2')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```

```python
# Pair plot:
numerical_cols = ['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']
sns.pairplot(df[numerical_cols].dropna())
plt.suptitle("Pair Plot of Titanic Numerical Features", y=1.02)
plt.show()
```

```python
# Grouped Bar Chart:
grouped = df.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack()
grouped.plot(kind='bar', stacked=False, colormap='Set3')
plt.title('Survival Rate by Class and Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
